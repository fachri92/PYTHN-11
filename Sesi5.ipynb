### Combining Dataset

```python
df8 = pd.DataFrame({'Fisika': [70,80,90,30], 'Matematika': [65,85,75,40], 'Biologi': [55, 45, 75, 85]},
                   index=['Andi', 'Budi', 'Caca', 'Dedi'])
df8
```

```python
df9 = pd.DataFrame({'Fisika': [70,80,90,30], 'Matematika': [65,85,75,40], 'Biologi': [55, 45, 75, 85]},
                   index=['Rosa', 'Ahmad', 'Tukiyem', 'Jajang'])
df9
```

```python
#merge
countries = pd.DataFrame({'population' : [10,20,30],
                         'continent' : ['europe','asia','america'],
                         }, index=['Holland','Japan','Canada'])
countries2 = pd.DataFrame({'president_gender' : ['M','F','M'],
                         'Minister' : ['Yes','No','Yes'],
                         }, index=['Holland','Japan','Canada'])
```

```python
countries
```

```python
countries2
```

```python
pd.merge(countries,countries2,left_index=True,right_index=True)
```
