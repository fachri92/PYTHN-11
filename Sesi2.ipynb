# While Loop

```python
apel = 5
while apel>0:
    apel-=1
    print(apel)
```

```python
apel = 5
apel -=1
print(apel)
```

```python
apel = 5
while apel>0:
    apel-=1
    if apel ==1:
        print('woy tinggal 1 nih ati2 ya')
    else:
        print(apel)
```

```python
# break digunakan untuk memberhentikan loop
apel = 5
while apel>0:
    apel-=1
    if apel ==2:
        break
    print(apel)
print('Loop Ended')
```

```python
apel
```

```python
# continue digunakan untuk melanjukan loop berdasarkan kondisi
apel = 5
while apel>0:
    apel-=1
    if apel ==2:
        continue
    print(apel)
print('Loop Ended')
```

```python
# else
apel = 10
while apel>0:
    apel-=2
    print(apel)
else:
    print('loop ended')
```

```python
# break digunakan untuk memberhentikan loop
apel = 5
while apel>0:
    apel-=1
    print(apel)
    if apel==2:
        break
else:
    print('Loop Done')
```

## Nested If - Else & Loops

```python
#Nested IF
umur = 42
gender = 'M'
if umur <18:
    if gender == 'M':
        print('Son')
    else:
        print('Daughter')
if umur >=18 and umur <65:
    if gender == 'M':
        print('Dad')
    else:
        print('Mom')
```

```python
# For Loop
b = ['Budi','Joko','Anwar']
for i in b:
    print(i)
```

```python
# Nested For Loop
b = ['Budi','Joko','Anwar']
for i in b:
    for j in i :
        print(j)
```

```python
#Continue
for i in ['Aster','Jubah','Nambo','Hachi']:
    if 'b' in i:
        continue
    print(i)
```

```python
#Pass
for i in ['Aster','Jubah','Nambo','Hachi']:
    if 'b' in i:
        pass
    print(i)
```

```python
#Break
for i in ['Aster','Jubah','Nambo','Hachi']:
    if 'b' in i:
        break
    print(i)
```

* Loop itu keseluruhan iterasi objek 1 loop diatas 4 iterasi
* Iterasi adalah istilah untuk 1 putaran loop
* continue akan memberhentikan iterasi dan lanjut ke iterasi berikutnya
* break akan memberhentikan looping secara keseluruhan

```python
for i in ['Aster','Jubah','Nambo','Hachi'] :
    print(i)
else:
    print('Loop Done')
```
